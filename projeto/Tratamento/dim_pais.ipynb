```python
from sqlalchemy import create_engine, exc
import pandas as pd
```

```python
usuario = 'root'
senha = 'root'
host = 'localhost'
portorigem = 3306
portdestino = 3306
bancoorigem = 'vendas'
bancodestino = 'vendas_dimensional'
```

```python
try:
    engineori = create_engine(f"mysql+pymysql://{usuario}:{senha}@{host}:{portorigem}/{bancoorigem}")
    with engineori.connect() as connori:
        print(f"Conexão com o banco '{bancoorigem}' estabelecida com sucesso.")
except exc.SQLAlchemyError as e:
    print(f"Erro ao conectar ao banco '{bancoorigem}': {e}")
    exit()
```

```python
try:
    enginedes = create_engine(f"mysql+pymysql://{usuario}:{senha}@{host}:{portdestino}/{bancodestino}")
    with enginedes.connect() as conndes:
        print(f"Conexão com o banco '{bancodestino}' estabelecida com sucesso.")
except exc.SQLAlchemyError as e:
    print(f"Erro ao conectar ao banco '{bancodestino}': {e}")
    exit()
```

```python
sql = f"select id,sigla,sigla3,pais,nomeinternacional,continente,regiao from paises"
paises = pd.read_sql(sql,engineori)
paises = paises.reset_index()
paises.columns=['id','idtransacional','sigla','sigla3','pais','nomeinternacional','continente','regiao']
paises['id'] += 1
try:
    paises.to_sql('dim_pais',if_exists='append',index=False,con=enginedes)
    print("Dimensão pais carregada com sucesso.")
except exc.SQLAlchemyError as e:
        print(f"Erro ao carregar: {e}")
```
